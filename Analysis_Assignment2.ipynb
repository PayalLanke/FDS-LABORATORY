```python
import pandas as pd
import matplotlib.pyplot as plt

# sample dataset with repeated cities
data = {
    'Student': ['Payal', 'Gauri', 'Shraddha', 'Sneha', 'Fiza', 'Mitali'],
    'Course': ['Math', 'Physics', 'Chemistry', 'Math', 'Biology', 'Physics'],
    'City': ['Mumbai', 'Mumbai', 'Delhi', 'Delhi', 'Nashik', 'Mumbai'],  # Mumbai and Delhi repeated
    'Marks': [88, 92, 75, 85, 90, 78],
    'Attendance': [95, 88, 80, 92, 85, 90],
}

df = pd.DataFrame(data)
df.head()
```

```python
# QUALITATIVE ANALYSIS 
print("Qualitative Analysis:")
print(df['Course'].value_counts())      # Frequency of courses
print(df['City'].value_counts())        # Frequency of cities

# Bar chart for the cource distribution
df['Course'].value_counts().plot(kind='bar', color=['orange', 'pink', 'green', 'lightblue'])
plt.title("Course Distribution")
plt.xlabel("Course")
plt.ylabel("Count")
plt.show()

# Bar chart for City distribution
df['City'].value_counts().plot(kind='bar', color=['red', 'pink', 'yellow', 'magenta'])
plt.title("City Distribution")
plt.xlabel("City")
plt.ylabel("Count")
plt.show()
```

```python
# QUANTITATIVE ANALYSIS
print("\nQuantitative Analysis:")
print(df[['Marks', 'Attendance']].describe())  # Summary statistics

# Histogram for Marks
df['Marks'].plot(kind='hist', bins=5, color='pink', edgecolor='black')
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

# Scatter plot: Marks vs Attendance
plt.scatter(df['Marks'], df['Attendance'], color='green')
plt.title("Marks vs Attendance")
plt.xlabel("Marks")
plt.ylabel("Attendance")
plt.show()
```
