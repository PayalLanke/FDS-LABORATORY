```python
import pandas as pd 
#data loading
df=pd.read_csv("E:\FDS LABORATORY\student_dataset.csv")
print(df.head())
```

```python
df.isnull().sum() #Count of null values
df.fillna(df.mean(),inplace=True) #filling null values
df.dropna()
```

```python
df.isnull().sum()              # missing values
df.fillna(0, inplace=True)     # fill NaN with 0
df.fillna(df.mean(), inplace=True)  # fill NaN with mean
df.dropna()                    # remove rows with NaN
df.duplicated().sum()          # check duplicates
df.drop_duplicates(inplace=True)  # remove duplicates
```

```python
import pandas as pd
data={
    'Department':['IT','CS','IT','ECE','CS','ECE'],
    'Student':['Payal','Gauri','Shraddha','Sakshi','Saish','Sayli'],
    'Marks':[90,88,45,67,95,78],
    'Age':[21,22,20,21,23,22]
}
```

```python
#DataFrame in pandas
df=pd.DataFrame(data)
print(df)
```

```python
#Grouby usning Department And Marks
df.groupby('Department')['Marks'].mean()
```

```python
#Aggregate Functions unsing groupby 
df.groupby('Department')['Marks'].agg(['mean','max','min','count'])
```

```python
#group by multiple columns
df.groupby(['Department','Age'])['Marks'].mean()
```

```python
df['Marks'].plot(kind='hist') # histogram
```

```python
df.plot(x='Department', y='Marks', kind='scatter') # scatter plot
```
